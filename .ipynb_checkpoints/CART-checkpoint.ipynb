```python
# -*- coding: utf-8 -*-
# author:zhengmingzhang
from numpy import *
import numpy as np
import pandas as pd
import operator
import copy
dataSet = pd.read_csv('watermelon_2.csv', delimiter=",")
Attributes=((dataSet.columns).tolist())[1:]
dataSet=np.matrix(dataSet)
dataSet = dataSet[:,1:]
print("数据集的属性有： %s"%Attributes)
print("数据集：%s"%dataSet)
m,n=np.shape(dataSet)
d = np.arange(0,m,1) 
a = np.arange(0,n-1,1)
```

```python
def calcGini(d):
    # 计算基尼值
    # d为子样本集的索引
    numEntries = len(d)
    types = [] # 数据集中分类的个数
    count = {} # 用来看每种分类的样本个数
    for i in range(numEntries):
        # 得到给定数据集中所有分类，和各自的个数
        currentY = dataSet[d[i], n-1]
        if currentY not in count:
            count[currentY] = 1
            types.append(currentY)
        else:
            count[currentY] += 1
    gini = 1
    for j in range(len(types)):
        # 计算基尼值
        prob = count[types[j]] / numEntries
        gini -= prob * prob
    return gini

def calcGiniIndex(d, a):
    # 计算基尼指数，a为给定的属性
    numAll = len(d)
    types = []
    count = {}
    for i in range(numAll):
        # 得到给定属性的不同取值和各个取值的个数
        currentY = dataSet[d[i], a]
        if currentY not in count:
            count[currentY] = [d[i]]
            types.append(currentY)
        else:
            count[currentY].append(d[i])
    giniindex = 0
    for j in range(len(types)):
        prob = len(count[types[j]]) / numAll
        giniindex += prob * calcGini(count[types[j]])
    return giniindex, types
```

```python
def chooseBestFeature(d):
    # 选择最优划分
    # 基尼指数越低越好
    # d为子样本的索引
    optginiGain = 0
    for i in range(n-1):
        # 对每个属性都计算一下它的基尼指数，选择最小的作为划分属性
        giniGain, types = calcGiniIndex(d, i)
        #print( giniGain)
        if (giniGain < optginiGain) or (i == 0):
            optginiGain = giniGain
            opt_a = i
    giniGain, types = calcGiniIndex(d, opt_a)
    # print("当前划分属性为 %s"%Attributes[opt_a])
    return opt_a, types

def splitDataSet(d, opt_a, v):
    # 按照给定划分属性来划分数据集
    # v为特征的值
    retDataSet = []
    numAll = len(d)
    for i in range(numAll):
        if (dataSet[d[i], opt_a] == v):
            retDataSet.append(d[i])
    return retDataSet
```

```python
# 当遍历完所有划分数据集的属性后，类的标签并不唯一，那么需要使用一种多数表决法来决定该叶子节点的分类
def majorityCnt(classList):
    classCount = {}
    for vote in classList:
        if vote not in classCount.keys(): 
            classCount[vote] = 0
        classCount[vote] += 1
    sortedClassCount = sorted(classCount.items(), key=operator.itemgetter(1), reverse=True)
    return sortedClassCount[0][0]

```

```python
# 创建树
def featureAllSame(d, attribute):
    # 判断样本在所有属性上是否取值相同的函数
    num_d = len(d)
    num_a = len(attribute)
    data = dataSet[d, :]
    data = data[:, attribute]
    print(data)
    for i in range(num_d):
        if (i == 0):
            attriList = data[i, :]
        if (attriList.all() != data[i, :].all()):
            return False
        else:
            attriList = data[i, :]
    return True
# 定义节点
class Node(object):
    def __init__(self, title):
        self.title = title
        self.v = 1
        self.children = []
        self.ID = -1
        self.deep = 0
        
def TreeGenerate(d, attribute, tile):
    num = len(d)
    classList = []
    node = Node(tile)
    for i in range(num):
        class_i = dataSet[d[i],n-1]
        classList.append(class_i)
    # 递归的三个终止条件
    # 1.当前节点包含的样本全都属于同一个类别则无需划分
    if (classList.count(classList[0]) == len(classList)):
        # list.count()可快速判断一个列表中的值是否唯一
        node.v = classList[0]
        print(node.v)
        return node
    # 2.当前属性集为空，或样本在所有属性上取值都相同则使用多数表决
    if (attribute.all == None or featureAllSame(d, attribute) == True):
        node.v = majorityCnt(classList)
        print(node.v)
        return node
    # 选择最佳划分属性和划分属性中各个取值
    best_a, a_value = chooseBestFeature(d)
    node.v = Attributes[best_a]+"=?"
    print(node.v)
    for v in a_value:
        d_v = splitDataSet(d, best_a, v)
        # 3. 当前节点包含的样本集为空，则使用多数表决
        if d_v == []:
            node.v = majorityCnt(classList)
            print(node.v)
            return node
        else:
            attribute_new = copy.deepcopy(attribute)
            node.children.append(TreeGenerate(d_v, attribute_new, v))
    return node
myTree = TreeGenerate(d, a, myTree)   
```

```python
# 画出决策树
import matplotlib.pyplot as plt
plt.rcParams[u'font.sans-serif'] = ['simhei']
plt.rcParams['axes.unicode_minus'] = False
# 定义文本框和箭头
decisionNode = dict(boxstyle="sawtooth", fc="0.8")
leafNode = dict(boxstyle = "round4", fc="0.8")
arrow_args = dict(arrowstyle="<-")
def plotNode(nodeTxt,centerPt,parentPt,nodeType):    
    plt.annotate(nodeTxt,xy=parentPt,xycoords='axes fraction',xytext=centerPt,                                
                 textcoords='axes fraction',va="center",ha="center",bbox=nodeType,arrowprops=arrow_args)
# 为了画一棵决策树，我们需要有多少个叶节点，从而确定x轴的长度，还需要知道树的深度从而确定y的高度
# 计算树有多少个叶节点
def getNumLeafs(myTree, ID):
    if (myTree.v == "是") or (myTree.v == "否"):
        myTree.ID = ID
        ID += 1
        return ID
    for i in myTree.children:
        ID = getNumLeafs(i, ID)
    return ID
getNumLeafs(myTree, 0)
# 计算树的深度
def getTreeDeep(myTree,deep):
    myTree.deep = deep
    res = 0
    if (myTree.v == "是") or (myTree.v == "否"):
        res += 1
        return res,deep
    maxdeep = deep
    for i in myTree.children:
        e,d = getTreeDeep(i, deep+1)
        res += e
        if d > maxdeep: maxdeep = d
    return res, maxdeep

cnt,deep = getTreeDeep(myTree,0)
def dfsPlot(myTree):
    if myTree.ID==-1 and len(myTree.children) != 0:
        childrenPx=[]
        meanPx=0
        for i in myTree.children:
            cur=dfsPlot(i)
            meanPx+=cur
            childrenPx.append(cur)
        meanPx=meanPx/len(myTree.children)
        c=0
        for i in myTree.children:
            nodetype=leafNode
            if i.ID<0:nodetype=decisionNode
            plotNode(i.v,(childrenPx[c],0.9-i.deep*0.8/deep),(meanPx,0.9-myTree.deep*0.8/deep),nodetype)
            plt.text((childrenPx[c]+meanPx)/2,(0.9-i.deep*0.8/deep+0.9-myTree.deep*0.8/deep)/2,i.title)
            c+=1
        return meanPx
    else:
        return 0.1+myTree.ID*0.8/(cnt-1)

rootX=dfsPlot(myTree)

plotNode(myTree.v,(rootX,0.9),(rootX,0.9),decisionNode)

```
